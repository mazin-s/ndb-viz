```python
import glob
import plotly.express as px
```

```python
raw_files = glob.iglob("../../**/*", recursive=True)
files = [file[6:] for file in raw_files]
```

```python
# Make the dir tree
data = {}
for file in files:
    parts = file.split("/")
    last_data = data
    for part in parts:
        if part not in last_data:
            last_data[part] = {}
        last_data = last_data[part]
    
```

```python
counts = {}
def get_plotly_data(data: dict[str, any], parent_name: str, names: list['str'], parents: list['str']):
    global counts
    for key, val in data.items():
        counts[key] = counts.get(key, -1) + 1
        key = f"{key}__{'' if counts[key] <= 0 else counts[key]}"
        names.append(key)
        parents.append(parent_name)
        if val != {}:
            get_plotly_data(val, key, names, parents)

names = []
parents = []
get_plotly_data(data, "root", names, parents)
names.append("root")
parents.append("")
for ix in range(len(names)):
    print(names[ix], parents[ix])
print(len(names), len(set(names)))
```

```python
fig = px.treemap(
    names = names,
    parents = parents,
)
fig.update_traces(root_color="lightgrey")
fig.update_layout(margin = dict(t=50, l=25, r=25, b=25))
fig.show()
```
